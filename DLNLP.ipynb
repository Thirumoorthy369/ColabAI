```python
#pip install -u tensorflow
import tensorflow as tf
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype("float32") / 255.0  #flatten apply pandrathu
x_test = x_test.astype("float32") / 255.0
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

model = tf.keras. Sequential([
tf.keras.layers.Dense (128, activation='relu', input_shape=(784,)),#L1 layer has 128 layers
tf.keras.layers.Dropout(0.3),
tf.keras.layers.Dense (64, activation='relu'),# L2 has 64 layers
tf.keras.layers.Dropout(0.3),
tf.keras.layers.Dense (10, activation='softmax') # Output has 10 layers
])
#keras sequential is best one...
model.compile(optimizer='adam',
loss='sparse_categorical_crossentropy',
metrics=['accuracy']) #gradient decent as adam for reverse also

history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print (f"Test accuracy: {test_acc:.3f}")

pred = model.predict(x_test[:10])
print("Predicted:", np.argmax (pred, axis=1))
print("True :", y_test[:10])
```
